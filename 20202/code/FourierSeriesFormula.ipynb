```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda a: a**2
x = np.linspace(-np.pi, np.pi, 100)
y = f(x)
T = np.pi

a_n = lambda n: (4*T**2 * (-1)**(n))/(n*np.pi)**2
b_n = lambda n: n*0
a_0 = (2*np.pi**2) / 3

f_series = np.zeros(len(x))+ a_0/2

for n in range(1,10):
    f_series += a_n(n)*np.cos(n*x) + b_n(n)*np.sin(n*x)

plt.plot(x, y, label=r'$f(x)=x^4$')
plt.plot(x, f_series, label="fourier series")
plt.legend()
```

```python
def factorial(n):
    result = 1
    if n>0:
        for i in range(2,n+1):
            result *= i
    return result
```
