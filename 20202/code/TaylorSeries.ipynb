```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def factorial(n):
    result = 1
    if n > 1:
        for i in range(2, n+1):
            result *= i
    return result
```

```python
f = lambda a: a**2
x = np.linspace(0,10,100)
y=f(x)

plt.plot(x,y)
plt.yscale('log')
```

```python
f_series = lambda a: (2*a**3)/ factorial(3)

y_series = f_series(x)

plt.plot(x,y, label=r'$x^2$')
plt.plot(x,y_series, label=r'taylor n = 3')
```
