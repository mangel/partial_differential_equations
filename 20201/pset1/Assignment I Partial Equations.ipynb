$1.$ Reproduce in full detail a proof of the Taylor theorem

#### Taylor's Theorem
Suppose that $f$ is defined on some open interval I around a and suppose $f(N+1)(x)$ exists on this interval. Then for each $x \neq a$ in I there is a value z between x and a so that
$$f(x) = \sum_{n=0}^N {f^{(n)}(a)\over n!}\,(x-a)^n + 
  {f^{(N+1)}(z)\over (N+1)!}(x-a)^{N+1}.$$

#### Proof
##### (Based on https://www.whitman.edu/mathematics/calculus_online/section11.11.html)


We start by defining a function $F(t)$:
$$F(t)=\sum_{n=0}^N{f^{(n)}(t)\over n!}\,(x-t)^n + B(x-t)^{N+1}.$$
notice that a $B$ term has been added to be at the end, and t is between $a$ and $x$. If $t=a$, then
$$F(a)=\sum_{n=0}^N{f^{(n)}(a)\over n!}\,(x-a)^n + B(x-a)^{N+1}.$$
Which will be exactly to $f(x)$ with the subtle diference of the substitution of $B$:
$$f(x)=\sum_{n=0}^N{f^{(n)}(a)\over n!}\,(x-a)^n + B(x-a)^{N+1}.$$
by the theorem statement $x \neq a$, so we can solve for $B$, which is a "constant"-it depends on $x$ and $a$ but those are temporaily fixed. We have defined a function $F(t)$ with the property $F(a)=f(x)$ on the interval. Consider also $F(x)$: all terms with a positive power of $(x - t)$ become zero when we substitute $x$ for $t$, so we are left with $F(x) = f^{0}(x)/0! = f(x)$. So $F(t)$ is a function with the same value on the endpoints of the interval $[a,x]$. By Rolle's theorem, we know that there is a value $z \in (a,x)$ such that $F'(z)=0$, lets look at $F'(t)$
$$
\eqalign{
  F(t)=f(t)&+{f^{(1)}(t)\over 1!}(x-t)^1+{f^{(2)}(t)\over 2!}(x-t)^2+
  {f^{(3)}(t)\over 3!}(x-t)^3+\cdots\cr
  &+{f^{(N)}(t)\over N!}(x-t)^N+
  B(x-t)^{N+1}.\cr}
$$
taking the derivative:
$$
\eqalign{
  F'(t) = f'(t) &+ 
  \left({f^{(1)}(t)\over 1!}(x-t)^0(-1)+{f^{(2)}(t)\over
    1!}(x-t)^1\right)\cr
  &+\left({f^{(2)}(t)\over 1!}(x-t)^1(-1)+{f^{(3)}(t)\over
    2!}(x-t)^2\right)\cr
  &+\left({f^{(3)}(t)\over 2!}(x-t)^2(-1)+{f^{(4)}(t)\over
    3!}(x-t)^3\right)+â€¦+\cr
  &+\left({f^{(N)}(t)\over (N-1)!}(x-t)^{N-1}(-1)+{f^{(N+1)}(t)\over
    N!}(x-t)^N\right)\cr
  &+B(N+1)(x-t)^N(-1).\cr}
$$
Notice that most of the terms in the previous expression cancel leaving just
$$F'(t) = {f^{(N+1)}(t)\over N!}(x-t)^N+B(N+1)(x-t)^N(-1).$$
But by Rolle's theorem at some $z$, $F'(z) = 0$ so
$$\eqalign{
  0&={f^{(N+1)}(z)\over N!}(x-z)^N+B(N+1)(x-z)^N(-1)\cr
  B(N+1)(x-z)^N&={f^{(N+1)}(z)\over N!}(x-z)^N\cr
  B&={f^{(N+1)}(z)\over (N+1)!}.\cr
}$$
Which can be rewritten as
$$F(t)=\sum_{n=0}^N{f^{(n)}(t)\over n!}\,(x-t)^n + 
  {f^{(N+1)}(z)\over (N+1)!}(x-t)^{N+1}.$$
 but as we saw previously we have the property $F(a) = f(x)$, to finally obtain:
$$f(x)=\sum_{n=0}^N{f^{(n)}(a)\over n!}\,(x-a)^n + 
  {f^{(N+1)}(z)\over (N+1)!}(x-a)^{N+1}$$
$\square$

The methods of finite difference, are numerical analysis techniques that approximate the values of functions, these methods are used to obtain numerical approximations of the solutions of systems of differential equation in which an analytical solution is not possible.

The method is based on a central idea of Taylor's Theorem about approximation of a differentiable continous function on an interval and several methods exists (implicit, explicit, Crank-Nicolson, etc).
$$\lim_{h \to 0} = \frac{f(x + h) - f(x)}{h} = f'(x)$$
On this case (the difference quotient), uses the construction of the derivative to approximate the value of the function,
$$f(x)\approx \frac{f(x + h) - f(x)}{h}$$
Notice that we are going to obtain an approximation and $h$ will be a parameter to divide the space on which we are making the approximations.

In general Differential Equation problems have initial conditions, for example, if we define the differential equation $f'=f$, with an initial value condition $f(0)=1$, by using the previous approximation we have:

$$
\begin{align}
    f(0) &= 1,\\
    f(x) &= \frac{f(x+h) - f(x)}{h} \Rightarrow f(x+h) = (1 + h)f(x),\\
\end{align}
$$

Incrementally we can find an approximation to the original values of $f$. Then

$$
\begin{align}
    f(0 + h) &= f(h) = (1 + h) f(0) = (1 + h),\\
    f(2h) &= (1 + h)f(h) = (1 + h)(1 + h),\\
    &\dots
\end{align}
$$

We will denote the approximations of $f$ as $f_k \approx f(x_k)$, where $x_k = kh$ on which $k$ becomes now an step on which we are making the approximations. Finally we will obtain:

$$f_0 = 1$$
$$f_{k+1} = (1+h)f_k$$

$3.$ Develop a code in python to solve the harmonic oscillator equation by finite difference.

The simple harmonic oscillator (SHO) is a mass connected to some elastic object of negligible mass that is fixed at the other end and constrained so that it may only move in one dimension.

The equation that describes the model is:
$$\ddot{x} + \omega_0^2 x = 0$$

where $\omega_0^2 = \omega_0^2 = \frac{k}{m}$, $k$ is the stiffness of the elastic object, and $m$ is the mass of the object. We need to add two initial conditions:
$$
\begin{align}
    x(t = 0) = x_0 \\
    \dot{x}(t=0) = \dot{x}_0
\end{align}
$$

To solve this problem with finite differences, we need to discretize the problem:
$$
\left\{
    \begin{align}
        x^{n+1} = && (2 - \omega_0^2dt^2)x^n - x^{n-1} \\
        x^0 = && x_0 \\
        x^1 = && x_0 + \dot{x}_0 dt
    \end{align}
\right.
$$

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('bmh')

# This class contains and abstraction of the simulation process
class HarmonicOdeSolver:
    # Constructor
    # receives the approximation parameters
    # dt: change in time
    # x0: the initial condition
    # xd0: the initial condition for the derivative
    # omega_squared: constant of stiffness over the mass of the object
    def __init__(self, dt, x0, xd0, omega_squared):
        # Discretization of the problem
        "Inits the solver."
        self.dt = dt
        self.dt_squared = dt**2
        self.t = dt
        self.omega_squared = omega_squared
        self.x0 = x0
        self.xd0 = xd0
        self.x = [xd0 * dt + x0, x0]
    
    # the computation for x at step {n+1}
    def step(self):
        "Steps the solver."
        xt, xtm1 = self.x
        xtp1 = (2 - self.omega_squared * self.dt_squared) * xt - xtm1
        self.x = (xtp1, xt)
        self.t += self.dt
        
    # the recurrent computation until n desired state
    def step_until(self, tmax, snapshot_dt):
        "Steps the solver until a given time, returns snapshots."
        ts = [self.t]
        vals = [self.x[0]]
        niter = max(1, int(snapshot_dt // self.dt))
        while self.t < tmax:
            for _ in range(niter):
                self.step()
            vals.append(self.x[0])
            ts.append(self.t)
        return np.array(ts), np.array(vals)
```

```python
snapshot_dt = 0.15

solver = HarmonicOdeSolver(2e-1, 1, 0, 4)

ts, vals = solver.step_until(12, snapshot_dt)

plt.plot(ts, vals);
```
